```python
#課題
from urllib.parse import urljoin
import requests
from bs4 import BeautifulSoup

DOMAIN = 'https://hosh-it.github.io/scraping-practice/'
text_list = []

response = requests.get(DOMAIN)
soup = BeautifulSoup(response.text, 'html.parser')

anchor_tags = soup.find_all('a')

for anchor_tag in anchor_tags:
    link = anchor_tag.get('href')
    url_body = urljoin(DOMAIN, link)

    response_body = requests.get(url_body)
    soup_body = BeautifulSoup(response_body.text, 'html.parser')
    p_tags = soup_body.find_all('p')
    #複数の文字列を改行で結合 内包表記
    texts = ''
    texts = "\n".join(p_tag.get_text() for p_tag in p_tags)

    text_list.append(texts)
    print(texts)
```
