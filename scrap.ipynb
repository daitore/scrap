```python
import requests

print(requests.__version__)
```

```python
url = 'https://hosh-it.github.io/scraping-practice/practice.html'
response = requests.get(url)
response.text

```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.text)
print(soup.prettify())
```

タグ名でimgタグを取得する

```python
soup.find_all('img')
print(soup.find_all('img'))
```

```python
# クラス名でimgタグを取得する
img_tag = soup.find('img',class_ ="size-middle")
img_tags = soup.find_all('img',class_ ="size-middle")
print(img_tag)
print(img_tags)
```

```python
#id名でimgタグを取得する
img_tag = soup.find('img',id="small-size")
print(img_tag)
```

```python
img_tag.get('src')
```

```python
img_tags = soup.find_all('img')
img_src_list = []
for img_tag in img_tags:
    img_src_list.append(img_tag.get('src'))
print(img_src_list)
```

```python
a_tags = soup.find_all('a')
a_href_list = []
for a_tag in a_tags:
    print(a_tag)
    print(a_tag.get('href'))
    a_href_list.append(a_tag.get('href'))
print(a_href_list)
```

```python
#要素内の文字列取得
a_tag = soup.find('a')
a_tag.get_text()
```

```python
#要素内 全文字列取得
a_tags= soup.find_all('a')
a_text_list = []
for a_tag in a_tags:
    print(a_tag)
    print(a_tag.get_text())
    a_text_list.append(a_tag.get_text())
print(a_text_list)
```

```python
import pandas as pd
shop = {
    '商品名': ['商品A', '商品B','商品C'],
    '金額' : [1000, 2000, 3000]
}
pd.DataFrame(shop)
```

```python
a_tags = soup.find_all('a')
img_tags = soup.find_all('img')

news = []
for a_tag, img_tag in zip(a_tags, img_tags):
    news.append({
        '記事のタイトル': a_tag.get_text(strip=True),
        'リンク先': a_tag.get('href'),
        '画像のリンク': img_tag.get('src')
    })

df = pd.DataFrame(news)
```

```python
df.to_csv('news.csv', index=False)
```
